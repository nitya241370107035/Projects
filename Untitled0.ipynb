```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import random
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler


# 1. Generate Synthetic Signal (can be replaced with real signal)
np.random.seed(0)
time = np.arange(0, 500, 0.1)
signal = np.sin(0.02 * np.pi * time) + np.sin(0.04 * np.pi * time)

# 2. Introduce missing points (simulate signal loss)
missing_ratio = 0.15  # 15% signal missing
signal_with_missing = signal.copy()
missing_indices = sorted(random.sample(range(len(signal)), int(len(signal) * missing_ratio)))
signal_with_missing[missing_indices] = np.nan

# 3. Prepare data: Normalize + Fill NaNs for training
scaler = MinMaxScaler()
signal_reshaped = signal.reshape(-1, 1)
signal_scaled = scaler.fit_transform(signal_reshaped).flatten()

signal_with_missing_scaled = signal_scaled.copy()
mean_signal = np.nanmean(signal_scaled)
signal_with_missing_scaled[missing_indices] = mean_signal  # Fill NaNs temporarily

# 4. Create sliding window dataset
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

window_size = 20
X, y = create_sequences(signal_with_missing_scaled, window_size)
X = X.reshape((X.shape[0], X.shape[1], 1))

# 5. Train LSTM only on known data
non_missing_mask = [(i + window_size) not in missing_indices for i in range(len(y))]
X_train = X[non_missing_mask]
y_train = y[non_missing_mask]

# 6. Build LSTM model
model = Sequential([
    LSTM(64, input_shape=(window_size, 1), return_sequences=False),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
history = model.fit(X_train, y_train, epochs=30, batch_size=32, verbose=1)

# 7. Predict missing values
X_full = X
y_pred = model.predict(X_full).flatten()

# Reconstruct signal with predicted values
reconstructed_signal = signal_with_missing_scaled.copy()
for i in missing_indices:
    if i >= window_size:
        reconstructed_signal[i] = y_pred[i - window_size]

# Inverse scaling to get back to original scale
final_reconstructed_signal = scaler.inverse_transform(reconstructed_signal.reshape(-1, 1)).flatten()
original_signal = signal
imputed_signal = signal.copy()
imputed_signal[missing_indices] = final_reconstructed_signal[missing_indices]

# 8. Plot Results
plt.figure(figsize=(14, 6))
plt.plot(time, signal, label='Original Signal', color='gray', alpha=0.6)
plt.plot(time, imputed_signal, label='Predicted Signal (LSTM)', color='orange', linewidth=2)
plt.scatter(time[missing_indices], final_reconstructed_signal[missing_indices],
            color='red', label='Predicted Missing Points', zorder=5)
plt.title('Missing Signal Prediction using LSTM Deep Learning')
plt.xlabel('Time')
plt.ylabel('Signal')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 9. Calculate Error on Missing Points Only
true_missing = signal[missing_indices]
predicted_missing = final_reconstructed_signal[missing_indices]
mse = np.mean((true_missing - predicted_missing) ** 2)
print(f"🔍 MSE on Predicted Missing Points: {mse:.6f}")

```
